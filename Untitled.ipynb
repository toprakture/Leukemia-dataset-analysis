```python
D = []
N = []

D.extend(training_dataset_0_all)
D.extend(training_dataset_1_all)
D.extend(training_dataset_2_all)

N.extend(training_dataset_0_hem)
N.extend(training_dataset_1_hem)
N.extend(training_dataset_2_hem)

D = np.array(D)
N = np.array(N)

len(D), len(N)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 5, figsize = (20,20))

for i in tqdm(range(0, 5)):
    rand = np.random.randint(len(D))
    img  = imread(D[rand])
    img  = resize(img, (128,128))
    ax[i].imshow(img)
    ax[i].axis('off')
```

```python
Image = []
Label = []

for i in tqdm(range(0, len(D))):
    img = imread(D[i])
    img = resize(img, (128,128))
    Image.append(img)
    Label.append(1)
    
for i in tqdm(range(0, len(N))):
    img = imread(N[i])
    img = resize(img, (128,128))
    Image.append(img)
    Label.append(0)
    
Image = np.array(Image)
Label = np.array(Label)

Image.shape, Label.shape
```

```python
del D
del N
```

```python
Image, Label = shuffle(Image, Label, random_state = 42)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 5, figsize = (20,20))

for i in tqdm(range(0, 5)):
    rand = np.random.randint(len(Image))
    ax[i].imshow(Image[rand])
    ax[i].axis('off')
    a = Label[rand]
    if a == 1:
        ax[i].set_title('Diseased')
    else:
        ax[i].set_title('NonDiseased')
```
